# Find the value of $x$ at which $f(x)$ is minimum : 
1. Find $x$ analytically
2. Write the update equation of gradient descent
3. Find $x$ using gradient descent method 



**Example 1** : $f(x)$ = $x^{2}$ + $x$ + $2$

**Analytical** : $$\frac{d}{dx} f(x)=2x+1=0$$
$$\frac{d^{2}}{dx^{2}} f(x)=2~(Minima)$$
$$x=-\frac{1}{2}~(analytical~solution)$$

**Gradient Descent Update equation** : $$x_{init}=4$$
$$x_{updt}=x_{old}-\lambda (\frac{d}{dx} f(x)|x=x_{old})$$
$$x_{updt}=x_{old}-\lambda (2x_{old}+1)$$

**Gradient Descent Method** :

Follow the below steps and write your code in the block below

1. Generate $x$, 1000 data points from -10 to 10
2. Generate and Plot the function $f(x)$ = $x^{2}$ + $x$ + $2$
3. Initialize the starting point ($x_{init}$) and learning rate ($\lambda$)
4. Use Gradient descent algorithm to compute value of $x$ at which the function $f(x)$ is minimum
5. Also vary the learning rate and initialisation point and plot your observations

```
# example of plotting a gradient descent search on a one-dimensional function
from numpy import asarray
from numpy import arange
from numpy.random import rand
from matplotlib import pyplot

x = np.random.randint(-10,10,1000)
# objective function
def objective(x):
	return x*x*2.0 +x+2

# derivative of objective function
def derivative(x):
	return x * 2.0 +1

# gradient descent algorithm
def gradient_descent(objective, derivative, bounds, n_iter, lr):
	# track all solutions
	solutions, scores = list(), list()
	# generate an initial point
	solution = -10 
	# run the gradient descent
	for i in range(n_iter):
		# calculate gradient
		gradient = derivative(solution)
		# take a step
		solution = solution - lr * gradient #updating x value
		# evaluate candidate point
		solution_eval = objective(solution)
		# store solution
		solutions.append(solution)
		scores.append(solution_eval)
		# report progress
		print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
	return [solutions, scores]

# define range for input
bounds = asarray([[-10.0, 10.0]])
# define the total iterations
n_iter = 100
# define the step size
lr = 0.85
# perform the gradient descent search
solutions, scores = gradient_descent(objective, derivative, bounds, n_iter, lr)
# sample input range uniformly at 0.1 increments
inputs = np.linspace(-10,10,1000)
# compute targets
results = objective(inputs)
# create a line plot of input vs result
plt.figure(figsize=(10,5))
pyplot.plot(inputs, results)
# plot the solutions found
pyplot.plot(solutions, scores, '.-', color='red')
# show the plot
pyplot.show()
```

#Find the value of $x$ and $y$ at which $f(x,y)$ is minimum :



**Example 1** : $f(x,y)$ = $x^{2}+y^{2}+2x+2y$

**Gradient Descent Method** :

Follow the below steps and write your code in the block below

1. Generate $x$ and $y$, 1000 data points from -10 to 10
2. Generate and Plot the function $f(x,y)$ = $x^{2}+y^{2}+2x+2y$
3. Initialize the starting point ($x_{init}$,$y_{init}$) and learning rate ($\lambda$)
4. Use Gradient descent algorithm to compute value of $x$ and $y$ at which the function $f(x,y)$ is minimum
5. Also vary the learning rate and initialisation point and plot your observations

```
## Write your code here (Ignore the warning)
from sympy import *
from mpl_toolkits.mplot3d import Axes3D

x = np.random.randint(-10,10,1000)
y=np.random.randint(-10,10,1000)

x = Symbol('x')
y = Symbol('y')
z = Symbol('z')

f = x**2 + y**2 +2*x +2*y #defining objective function

# First partial derivative with respect to x
fpx = f.diff(x)

# First partial derivative with respect to y
fpy = f.diff(y)

# Gradient
grad = [fpx,fpy]

# Data
x_init = -10  #x
y_init = -10  #y
lr = .01
iterations = 0
check = 0
precision = 1/1000000
printData = True
maxIterations = 1000
solutions_x =[] #defining empty list
solutions_y =[]

while True:
    x_updt = x_init - lr*(fpx.subs(x,x_init).subs(y,y_init))
    y_updt = y_init - lr*(fpy.subs(y,y_init)).subs(x,x_init)
      
    solutions_x.append(x_updt)
    solutions_y.append(y_updt)
    

    #If the number of iterations goes up too much, maybe theta (and/or theta1)
    #is diverging! Let's stop the loop and try to understand.
    iterations += 1
    if iterations > maxIterations:
        print("Too many iterations. Adjust alpha and make sure that the function is convex!")
        printData = False
        break

    #If the value of theta changes less of a certain amount, our goal is met.
    if abs(x_updt-x_init) < precision and abs(y_updt-y_init) < precision:
        break

    #Simultaneous update
    x_init = x_updt
    y_init = y_updt

if printData:
    print("The function "+str(f)+" converges to a minimum")
    print("Number of iterations:",iterations,sep=" ")
    print("x_value =",x_updt,sep=" ")
    print("y_value =",y_updt,sep=" ")


#plotting the graph
fig = plt.figure(figsize=(10,15))
#ax.clabel(CS, inline=1, fontsize=10) 
x, y = np.linspace(-10,10,1000), np.linspace(-10,10,1000)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2 +2*X +2*Y
ax = fig.add_subplot(311, projection='3d')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.plot_surface(X,Y,Z, alpha=0.5, rstride=100, cstride=100)
ax = fig.add_subplot(312,projection ='3d')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.contour3D(X,Y,Z,50,isocline=False)
ax = fig.add_subplot(313)
ax.contour(X,Y,Z,200, isocline=True)
for i in range(len(solutions_x)-1):
  ax.annotate('', xy=(solutions_x[i+1],solutions_y[i+1]), xytext= (solutions_x[i],solutions_y[i]) ,arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},va='center', ha='center')

plt.show()
```
